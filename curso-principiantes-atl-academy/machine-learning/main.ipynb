### Set de datos de deportes
pip install scikit-learn

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
#Cargar los datos desde el archivo CSV
dataset = pd.read_csv('deportes.csv')
#Separa las caracteristicas y variables objetivo
x = dataset[['genero','edad']]
y = dataset['deporte']
#Creando modelo de árbol de decisión
modelo = DecisionTreeClassifier()
#Entrenar modelo
modelo.fit(x.values, y)
#Hacer predicciones
predictions  = modelo.predict([ 
    [1,24], [2,33]
    ])
print(predictions)
```
